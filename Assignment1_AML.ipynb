## KNN Analysis  and Accuracy for the Artififcial Dataset

```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate artificial dataset with three class centers using make_blobs
Artificial_dataset = [[-5, -2], [3, 9], [10, -3]] 
# Updated centers for three classes
n_classes = len(Artificial_dataset)
data, labels = make_blobs(n_samples=150, centers=np.array(Artificial_dataset), random_state=42)

# Split the data into 80% training and 20% testing
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
# Create and fit a K-Nearest Neighbors classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(train_data, train_labels)

# Predict on the test data
predicted_labels = knn.predict(test_data)

# Calculate and print the accuracy score
accuracy = accuracy_score(test_labels, predicted_labels)
print("Accuracy on the test data:", accuracy)
```

```python
# Plot the results
# Plot training data
plt.figure(figsize=(8, 6))
plt.scatter(train_data[:, 0], train_data[:, 1], c=train_labels, cmap='viridis', marker='o', edgecolor='k', label='Train Data')

# Plot test data with predictions
plt.scatter(test_data[:, 0], test_data[:, 1], c=predicted_labels, cmap='coolwarm', marker='s', edgecolor='k', label='Test Data')

# Label the plot
plt.title('KNN Classification Results on Artificial Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
